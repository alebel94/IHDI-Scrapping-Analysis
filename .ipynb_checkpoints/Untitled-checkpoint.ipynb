```python
wikitable_class = soup.find_all("table", {"class": "wikitable"})[0]
tbody_class = list(wikitable_class.children)[1]
td_class_titles = list(tbody_class.children)[0]
td_class_titles
```

```python
'''
function removed_tag takes in:
parameter "soup" from the soup from URL provided.
parameter "tag" to find the specific tag in the soup.
parameter "regex" (str or list of str) of the expressions to be removed from those tags
'''

def removed_tag(soup=soup, tag= '', regex = ''):
    import re
    text_corpus = soup.find_all(tag) 
    final_text = []
    
    if type(regex) == str:
        regex_in = regex
    elif type(regex) == list:
        regex_in = '|'.join(regex)
    else:
        raise ValueError('Use only a regex string or list of regex strings')

    for node in text_corpus:
        text = node.get_text()
        text= re.sub(regex_in, '', text)
        final_text.append(text)
    return final_text

filtered_titles = removed_tag(td_class_titles, 'td' , ["\[[^\]]*\]", '\n'])
filtered_titles
```

```python
titles = {}
def make_titles(arr):
    j = 0
    for i, val in enumerate(arr, 1):
        if val != '\n':
            j += 1
            titles[f'Title_{j}'] = val
    return titles

make_titles(filtered_titles)
```
