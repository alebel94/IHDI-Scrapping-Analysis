```python
# Gender Inequality Index (GII)
gii = pd.read_csv('csv_final_project/Gender_Inequality_Index/Gender_Inequality_Index_2019.csv', sep = ',').iloc[:, 1:]
gii.rename(columns={'GII Rank':'GII_Rank', 'GII Value': 'GII_Value'}, inplace=True)
gii_condensed = gii[['Country', 'GII_Value', 'GII_Rank']]

df = gii_condensed
df['Country'] = df['Country'].replace(['Syrian Arab Republic', 'Viet Nam', 'Macedonia', 'Russian Federation', 'Czechia'],['Syria', 'Vietnam', 'Republic of North Macedonia', 'Russia', 'Czech Republic'])


gii_condensed
```

```python
#Social Progress Index (SPI)
spi = pd.read_csv('csv_final_project/Social_Progress_Index/Social_Progress_Index_2020_report.csv', sep = ',').iloc[:, 1:]
spi.rename(columns={'Rank':'SPI_Rank', 'Score': 'SPI_Score'}, inplace=True)
spi_condensed = spi[['Country', 'SPI_Score', 'SPI_Rank']]

df = spi_condensed
df['Country'] = df['Country'].replace(['Congo Republic of', 'Congo Democratic Republic of', "Cote D'Ivoire", 'Gambia The'],['Republic of the Congo', 'Democratic Republic of the Congo', "Côte d'Ivoire", 'Gambia'])

spi_condensed
```

```python
#World Happiness Report (WHR)
whr = pd.read_csv('csv_final_project/World_Happiness_Report/World_Happiness_Report_2020(2017-2019avg).csv', sep = ',').iloc[:, 1:]
whr.rename(columns={'Overall rank':'WHR_Rank', 'Country or region':'Country', 'Score':'WHR_Score'}, inplace=True)
whr_condensed = whr[['Country', 'WHR_Score', 'WHR_Rank']]

df = whr_condensed
df['Country'] = df['Country'].replace(['Congo (Brazzaville)', 'Congo (Kinshasa)', 'Ivory Coast', 'Macedonia'],['Republic of the Congo', 'Democratic Republic of the Congo',"Côte d'Ivoire", 'Republic of North Macedonia'])

whr_condensed



```

```python
pd.options.display.max_rows = None
full_table = ihdi_condensed.merge(gii_condensed, on = 'Country', how = 'outer')
full_table_dropped = full_table.drop_duplicates(subset = 'Country').reset_index(drop = True)

full_table = full_table.merge(spi_condensed, on = 'Country', how = 'outer')
full_table_dropped = full_table.drop_duplicates(subset = 'Country').reset_index(drop = True)

full_table = full_table.merge(whr_condensed, on = 'Country', how = 'outer')
full_table_dropped = full_table.drop_duplicates(subset = 'Country').reset_index(drop = True)

full_table_dropped = full_table_dropped[['Country', 'HDI_Rank', 'SPI_Rank', 'GII_Rank', 'WHR_Rank', 'HDI', 'IHDI', 'SPI_Score', 'GII_Value', 'WHR_Score']]

full_table_dropped = full_table_dropped.replace(float('NaN'), '-1')
full_table_dropped = full_table_dropped.astype({'HDI_Rank': 'int64', 'SPI_Rank': 'int64', 'GII_Rank': 'int64', 'WHR_Rank': 'int64', 'HDI': 'float', 'IHDI': 'float', 'SPI_Score': 'float', 'GII_Value': 'float', 'WHR_Score': 'float'})
full_table_dropped = full_table_dropped.replace(-1, np.nan)

main_table = full_table_dropped.dropna(how='any').reset_index(drop = True)

main_table
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(20,5))
plt.subplots_adjust(hspace=0.4)

corr_to_ihdi = pd.Series(dtype='float64')
for col in main_table.columns.values[7:]:
        corr_to_ihdi[col] = main_table['HDI'].corr(main_table[col])
abs_corr_to_ihdi = corr_to_ihdi.abs().sort_values(ascending=False)
corr_to_ihdi = corr_to_ihdi.loc[abs_corr_to_ihdi.index]
corr_to_ihdi

for i in range(3):
        sns.regplot(x=corr_to_ihdi.index.values[i], y='IHDI', data=main_table, ax=axes[i], fit_reg=False, marker='.')
        title = 'correlation='+str(corr_to_ihdi[i])
        axes[i].set_title(title)
plt.show()
```

```python
for i, corr_to_ihdi in enumerate(corr_to_ihdi):
    sns.regplot(x=corr_to_ihdi[i], y='IHDI', data=main_table, ax=axes[i,j], fit_reg=False, marker='.')
    title = 'correlation='+str(corr_to_ihdi[i])
    axes[i,j].set_title(title)
axes[1,2].set_xlim(0,102)
```
